```python
import pandas as pd
movies = pd.read_csv('https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv')
```

1. trovare la parola Porno nei nomi dei film

```python
movies[movies["Film"].str.contains("Porno")]
```

2. trovare la parola de nei nomi dei film

```python
movies[movies["Film"].str.contains("The")]
```

3. trovare i film che cominciano con la t

```python
movies[movies["Film"].str.startswith("T")]
```

4. trovare i film che finiscono in on

```python
movies[movies["Film"].str.endswith("on")]
```

5. visualizzare i titoli dei film in ordine alfabetico

```python
movies.sort_values(by=['Film'])
```

6. visualizzare i titoli dei film da z-a

```python
movies.sort_values(by=["Film"], ascending = False)
```

7. ordinare in ordine alfetico sul genere e poi all'interno del genere, ordinare i film

```python
movies.sort_values(by=["Genre", "Film"])
```
