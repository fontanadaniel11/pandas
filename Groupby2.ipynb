```python
import pandas as pd 
```

```python
squadre = pd.read_excel("https://raw.githubusercontent.com/wtitze/3E/main/Teams.xlsx")
```

```python
squadre
```

1. quante sono le squadre italiane?

```python
squadre[squadre["NOC"] == "Italy"]["NOC"].count()
```

2. quante sono le squadre tedesche?

```python
squadre[squadre["NOC"] == "Germany"]["NOC"].count()
```

3. quante sono le squadre di ogni nazione?

```python
squadregroup = squadre.groupby("NOC")[["NOC"]].count()
```

```python
squadregroup
```

```python
#visualizzare le nazioni che hanno più di 10 squadre
squadregroup[squadregroup["NOC"] > 10]
```

OLAP: si analizzano le informazioni estraendone i valori

```python
squadregroup = squadre.groupby("NOC").count().reset_index() #reset_index serve per inserire la colonna dell'indice all'inizio
squadregroup
```
