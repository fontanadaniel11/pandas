5. visualizzare i titolo e l’anno dei film che si trovano nelle posizioni 23, 29 e 45

9. visualizzare l’ “Audience score %” medio dei film del genere “Drama”


```python
movies[movies["Genre"] == "Drama"]["Audience score %"].mean()
```

10. diminuire del 10% il valore delle recensioni presenti nel campo “Audience score %”


```python
movies["Audience score %"] = movies["Audience score %"] - (movies["Audience score %"]/100*10)
movies.head()
```

```python
movies["Audience score %"] = movies["Audience score %"] * 0.9
```

```python
#aumentare del 10%
movies["Audience score %"] = movies["Audience score %"] * 1.1
```
