15. trovare il prezzo medio dei vini italiani, confrontarlo con il prezzo medio dei vini francesi e dire quali costano di più

```python
#metodo giusto
print("i vini italiani costano di più" if  df[df["country"] == "Italy"]["price"].mean() > df[df["country"] == "France"]["price"].mean() else "i vini francesi costano di più" )
```

```python
print(df[df["country"] == "Italy"]["price"].mean())
print(df[df["country"] == "France"]["price"].mean())
if df[df["country"] == "Italy"]["price"].mean() > df[df["country"] == "France"]["price"].mean():
  print("i vini italiani costano di più")
else:
  print("i vini francesi costano di più")
```
